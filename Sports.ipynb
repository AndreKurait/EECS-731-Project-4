```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
df_games = pd.read_csv('./data/nfl_games.csv')
```

```python
df_games = df_games.dropna()
df_games
```

We are now going to label encode the string entries

```python
# I will use a label encoder to encode the teams
le_team = preprocessing.LabelEncoder()
le_team.fit(df_games['team1'].append(df_games['team2']))
df_games['team1'] = le_team.transform(df_games['team1'])
df_games['team2'] = le_team.transform(df_games['team2'])
del df_games['date']
df_games.head()
```

```python
# Creating test and train split
scores = pd.DataFrame(df_games['score1'])
scores['score2'] = df_games['score2']
features = df_games[['season', 'neutral','playoff','team1','team2', 'elo1', 'elo2', 'elo_prob1']]
X_train1, X_test1, y_train1, y_test1 = train_test_split(features, scores, test_size=0.1)
```

```python
model1 = RandomForestRegressor(n_estimators=100)
model1.fit(X_train1, y_train1)
y_predict1 = model1.predict(X_test1)
r2_score(y_test1,y_predict1)
```

```python
model2 = MLPRegressor()
model2.fit(X_train1, y_train1)
y_predict2 = model2.predict(X_test1)
r2_score(y_test1,y_predict2)
```

Overall, these results are not great, but predicting scores are very hard.  We can see that the Random Forrest Regressor beat the MLP Regressor. We can look at predicting the winner and see our accuracy with that.

```python
winner_real = (y_test1['score1'] > y_test1['score2']).astype(int)
winner_predict = (y_predict1[:,0] > y_predict1[:,1]).astype(int)
accuracy_score(winner_real,winner_predict)
```

This shows that we can do much better at predicting who wins as opposed to predicting the score that occurs. At-least we are directionally correct on the score prediction most of the time. Plotting our random forrest regressor model vs the expected yeilds the following graph

```python
plt.scatter(y_test1, y_predict1, alpha=0.25)
plt.xlabel("Expected Score")
plt.ylabel("Model Prediction RFR")
plt.show()
```

This compared too the MLP regressor shows visually how much better our RFR is.
